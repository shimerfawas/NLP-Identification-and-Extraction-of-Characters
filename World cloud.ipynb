```python
import urllib.request

response = urllib.request.urlopen('https://whc.unesco.org/en/list/450/')

html = response.read()

print(html)
print(len(html))
```

```python
import re

html_content = html.decode('utf-8')
#we use non word characters to break
#W+ means 0 or more
tokens = re.split('\W+',html_content)

print(tokens)
```

```python
#using a library to remove html tags
#cleaned tokenized version of html content

from bs4 import BeautifulSoup
import nltk

html_content = html.decode('utf-8')

soup = BeautifulSoup(html_content,'html.parser')
clean_text = soup.get_text()
#print(clean_text)

tokens = [tok for tok in clean_text.split()]
print(tokens)
```

```python
#2nd way using library
import nltk

Freq_dist_nltk = nltk.FreqDist(tokens)
print(Freq_dist_nltk)

for k,v in Freq_dist_nltk.items():
  print(k,v)
```

```python
from nltk.corpus import stopwords
stopword_list = set(stopwords.words('english'))

clean_tokens = [tok for tok in tokens if len(tok.lower()) > 1 and (tok.lower() not in stopword_list)]
print(len(clean_tokens))
print(tokens)

```

```python
import nltk
Freq_dist_nltk = nltk.FreqDist(clean_tokens)
Freq_dist_nltk.plot(50,cumulative=False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_freq_dict = {word: freq for word, freq in Freq_dist_nltk.items()}
print(word_freq_dict)

word_cloud = WordCloud(width =800,
                       height= 800,
                       background_color='white',
                       min_font_size = 10)
word_cloud.generate_from_frequencies(word_freq_dict)
plt.figure()
plt.imshow(word_cloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
